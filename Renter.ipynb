```python
class RentalCalculator:
    def __init__(self):
        self.expenses = {}
        self.income = {}
        self.startup = {}
        
        self.total_expenses = sum(self.expenses.values())
        self.total_startup = sum(self.startup.values())
        self.total_income = sum(self.income.values())
        
        self.cash_flow = self.total_income - self.total_expenses
        self.total_investment = sum(self.startup.values())
        self.annual_cash_flow = self.cash_flow * 12
        if self.total_investment != 0:
            self.roi = (self.annual_cash_flow / self.total_investment) * 100
        else:
            self.roi = 0
        self.perc_roi = str(round(self.roi, 2)) + "%"

    def show_values(self):
        section = input('Expenses/Income/Initial Costs/Exit')
        if section.lower() == 'expenses':
            print(self.expenses)
            print(f'Total Monthly Expenses: {self.total_expenses}')
        elif section.lower() == 'income':
            print(self.income)
        elif 'initial' in section.lower():
            print(self.startup)
        elif section.lower() == 'exit':
            return
        else:
            print('Invalid section choice.')

    def adjust_values(self):
        section = input('Expenses/Income/Initial Costs/Exit')
        if section.lower() == 'expenses':
            print(self.expenses)
            value_choice = input('Which Value?')
            if value_choice in self.expenses:
                new_value = float(input('Enter The New Value: '))
                self.expenses[value_choice] = new_value
                print('Value Updated.')
        elif section.lower() == 'income':
            print(self.income)
            value_choice = input('Which Value?')
            if value_choice in self.income:
                new_value = float(input('Enter The New Value: '))
                self.income[value_choice] = new_value
                print('Value Updated.')
        elif 'initial' in section.lower():
            print(self.startup)
            value_choice = input('Which Value?')
            if value_choice in self.startup:
                new_value = float(input('Enter The New Value: '))
                self.startup[value_choice] = new_value
                print('Value Updated.')
        elif section.lower() == 'exit':
            return
        else:
            print('Invalid section choice.')

    def add_keys(self):
        dict_select = input('Expenses/Income/Initial Costs/Exit')
        if dict_select.lower() == 'expenses':
            key = input('Enter New Expense: ')
            value = float(input('Enter The Expense Value: '))
            self.expenses[key] = value
        elif dict_select.lower() == 'income':
            key = input('Enter New Income Source: ')
            value = float(input('Enter The New Source Value: '))
            self.income[key] = value
        elif 'initial' in dict_select.lower():
            key = input('Enter New Startup Cost: ')
            value = float(input('Enter The New Cost Value: '))
            self.startup[key] = value
        elif dict_select.lower() == 'exit':
            return
        else:
            print('Invalid section choice.')


display = RentalCalculator()
while True:
    option = input('Show Values/Adjust Values/Add New Items/Quit')
    if 'show' in option.lower():
        display.show_values()
    elif 'adjust' in option.lower():
        display.adjust_values()
    elif 'add' in option.lower():
        display.add_keys()
    elif 'quit' in option.lower():
        break
```
